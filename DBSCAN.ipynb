```python
from sklearn.datasets import make_moons
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
```

```python
X, _ = make_moons(n_samples = 300, noise = 0.05 , random_state = 0)
```

```python
X_scaled = StandardScaler().fit_transform(X)
```

```python
print(X)
```

```python
X_scaled
```

```python
db = DBSCAN(eps = 0.24, min_samples = 5)
```

```python
labels = db.fit_predict(X_scaled)
```

```python
labels
```

```python
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels, cmap='plasma')
plt.title("DBSCAN Clustering Result (with Outlier Detection)")
plt.show()
```

```python
from sklearn.neighbors import NearestNeighbors
import numpy as np

neighbors = NearestNeighbors(n_neighbors=5)
neighbors_fit = neighbors.fit(X_scaled)
distances, indices = neighbors_fit.kneighbors(X_scaled)

distances = np.sort(distances[:, 4])
plt.plot(distances)
plt.title("K-distance Graph (Look for Elbow)")
plt.show()
```

```python
from sklearn.metrics import silhouette_score

score = silhouette_score(X_scaled, labels)
print("Silhouette Score:", score)
```

```python
from sklearn.neighbors import NearestNeighbors
import matplotlib.pyplot as plt
import numpy as np

neighbors = NearestNeighbors(n_neighbors=5)
neighbors_fit = neighbors.fit(X_scaled)
distances, indices = neighbors_fit.kneighbors(X_scaled)

k_distances = np.sort(distances[:, 3]) 
plt.plot(k_distances)
plt.title("K-Distance Graph (elbow = good eps)")
plt.xlabel("Points sorted by distance")
plt.ylabel("4th Nearest Neighbor Distance")
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score

for min_samples in [3, 5, 7]:
    for eps in [0.2, 0.3, 0.4, 0.5]:
        model = DBSCAN(eps=eps, min_samples=min_samples)
        labels = model.fit_predict(X_scaled)
        n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
        if n_clusters > 1:
            score = silhouette_score(X_scaled, labels)
            print(f"eps={eps}, min_samples={min_samples}, clusters={n_clusters}, silhouette={score:.3f}")
```

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score

best_score = -1
best_model = None

for eps in np.arange(0.2, 0.6, 0.05):
    for min_samples in range(3, 8):
        db = DBSCAN(eps=eps, min_samples=min_samples)
        labels = db.fit_predict(X_scaled)
        if len(set(labels)) > 1 and -1 not in set(labels):  
            score = silhouette_score(X_scaled, labels)
            if score > best_score:
                best_score = score
                best_model = db

print(" Best Model  eps =", best_model.eps, "| min_samples =", best_model.min_samples, "| silhouette score =", best_score)
```

```python
from sklearn.cluster import OPTICS

optics = OPTICS(min_samples=5)
labels_optics = optics.fit_predict(X)
```

```python
def plot_optics_simple(X, optics_model, labels):
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    unique_labels = set(labels)
    colors = plt.cm.Set1(np.linspace(0, 1, len(unique_labels)))
    
    for k, col in zip(unique_labels, colors):
        if k == -1:
            
            class_member_mask = (labels == k)
            xy = X[class_member_mask]
            ax1.scatter(xy[:, 0], xy[:, 1], c='red', marker='x', 
                       s=100, alpha=0.8, label='Noise', linewidths=2)
        else:
            class_member_mask = (labels == k)
            xy = X[class_member_mask]
            ax1.scatter(xy[:, 0], xy[:, 1], c=[col], s=60, 
                       alpha=0.8, label=f'Cluster {k}')
    
    ax1.set_title('OPTICS Clustering Results', fontsize=14)
    ax1.set_xlabel('Feature 1')
    ax1.set_ylabel('Feature 2')
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    space = np.arange(len(X))
    reachability = optics_model.reachability_[optics_model.ordering_]
    labels_ordered = labels[optics_model.ordering_]
    
    colors_reach = []
    for label in labels_ordered:
        if label == -1:
            colors_reach.append('red')
        else:
            colors_reach.append(plt.cm.Set1(label / max(labels) if max(labels) > 0 else 0))
    
    ax2.bar(space, reachability, color=colors_reach, alpha=0.8, width=1.0)
    ax2.set_title('Reachability Plot', fontsize=14)
    ax2.set_xlabel('Points (ordered by OPTICS)')
    ax2.set_ylabel('Reachability Distance')
    ax2.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
    n_noise = list(labels).count(-1)
    print(f"OPTICS Results:")
    print(f"- Found {n_clusters} clusters")
    print(f"- {n_noise} noise points")
    
    return fig

plot_optics_simple(X, optics, labels_optics)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize

def plot_optics_simple(X, optics_model, labels, save_path=None):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

    unique_labels = sorted(set(labels))
    n_clusters = len([k for k in unique_labels if k != -1])
    n_noise = list(labels).count(-1)

    # Normalize cluster IDs for consistent colormap usage
    cluster_labels = [k for k in unique_labels if k != -1]
    color_norm = Normalize(vmin=0, vmax=max(cluster_labels) if cluster_labels else 1)
    colormap = plt.cm.get_cmap('Set1', len(cluster_labels))

    # Plot 1: Clustered points
    for k in unique_labels:
        mask = (labels == k)
        xy = X[mask]
        if k == -1:
            ax1.scatter(xy[:, 0], xy[:, 1], c='red', marker='x', s=100,
                        alpha=0.8, label='Noise', linewidths=2)
        else:
            ax1.scatter(xy[:, 0], xy[:, 1],
                        c=[colormap(color_norm(k))], s=60,
                        alpha=0.8, label=f'Cluster {k}')

    ax1.set_title(f'OPTICS Clustering Results', fontsize=14)
    ax1.set_xlabel('Feature 1')
    ax1.set_ylabel('Feature 2')
    ax1.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    ax1.grid(True, alpha=0.3)

    # Plot 2: Reachability plot
    ordering = optics_model.ordering_
    reachability = optics_model.reachability_[ordering]
    labels_ordered = labels[ordering]

    reach_colors = []
    for label in labels_ordered:
        if label == -1:
            reach_colors.append('red')
        else:
            reach_colors.append(colormap(color_norm(label)))

    ax2.bar(np.arange(len(X)), reachability,
            color=reach_colors, alpha=0.8, width=1.0)
    ax2.set_title('Reachability Plot', fontsize=14)
    ax2.set_xlabel('Points (ordered by OPTICS)')
    ax2.set_ylabel('Reachability Distance')
    ax2.grid(True, alpha=0.3)

    plt.tight_layout()

    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f" Plot saved to {save_path}")

    plt.show()

    print("OPTICS Results Summary:")
    print(f" Clusters found : {n_clusters}")
    print(f" Noise points   : {n_noise}")

    return fig
```

```python
plot_optics_simple(X, optics, labels_optics, save_path="optics_result.png")
```
