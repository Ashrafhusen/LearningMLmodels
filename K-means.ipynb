```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
X, _ = make_blobs(n_samples = 400, centers = 3, cluster_std = 0.60, random_state = 0)
```

```python
X
```

```python
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
kmeans = KMeans(n_clusters =4, random_state = 42, n_init=10)
```

```python
y_kmeans = kmeans.fit_predict(X_scaled)
```

```python
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y_kmeans, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', marker='X')
plt.title("K-Means Clustering Result")
plt.show()
```

```python
import numpy as np


new_data = np.array([
    [2, 3],
    [0, 5],
    [-4, -3],
    [6, 0]
])


new_data_scaled = scaler.transform(new_data)

predicted_clusters = kmeans.predict(new_data_scaled)

for i, point in enumerate(new_data):
    print(f"Point {point} is predicted to be in cluster {predicted_clusters[i]}")
```

```python
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y_kmeans, cmap='viridis', alpha=0.6)

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            s=300, c='red', marker='X', label='Centers')

plt.scatter(new_data_scaled[:, 0], new_data_scaled[:, 1],
            c='black', marker='*', s=200, label='New Data')

plt.title("K-Means with New Data Points")
plt.legend()
plt.show()
```

```python
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    sse.append(kmeans.inertia_)

plt.plot(range(1, 11), sse, marker='o')
plt.xlabel("K")
plt.ylabel("Sum of Squared Errors (Inertia)")
plt.title("Elbow Method for Optimal K")
plt.show()
```
